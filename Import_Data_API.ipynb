```python
from google.colab import drive
import pandas as pd
import requests
import json
import csv
import time
import os

drive.mount('/content/gdrive',force_remount=True)

path = '/content/gdrive/My Drive/Data Deli/'

timestr = time.strftime("%Y%m%d-%H:%M.csv")

filename = os.path.join(path,timestr)

headers = {
    'accept' : 'application/json',
    'Content-type' : 'application/json'
}
response = requests.get("https://api.thingspeak.com/channels/1952002/feeds.json?")
myjson = response.json()
ourdata = []
csvheader=['Created At','entry id','Temperatur','Kelembapan Relatif','PM2.5','CO2','Tekanan udara','Temperatur SHT31','Humidity SHT31']


#7 jam Lebih Awal
for x in myjson ['feeds']:
  listing = [x['created_at'],x['entry_id'],x['field1'],x['field2'],x['field3'],x['field4'],x['field5'],x['field6'],x['field7']]
  ourdata.append(listing)

with open(filename,'w',encoding='UTF8',newline='') as f:
  writer = csv.writer(f)

  writer.writerow(csvheader)
  writer.writerows(ourdata)

print(ourdata)


```
